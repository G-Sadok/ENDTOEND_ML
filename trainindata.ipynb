```python
import matplotlib.pyplot as plt,pickle
import pandas as pd

# from sklearn.model_selection import train_test_split
# from sklearn.metrics import classification_report,confusion_matrix
# from sklearn import metrics
from sklearn.svm import LinearSVC
from sklearn.ensemble import BaggingClassifier
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
alldata= pd.read_csv('imdb.csv',encoding="latin1")

alldata=alldata[alldata['label']!='unsup']
trainsplit=alldata[alldata['type']=='train']
testsplit=alldata[alldata['type']=='test']
```

```python
x_train=trainsplit['review']
y_train=trainsplit['label']

x_test=testsplit['review']
y_test=testsplit['label']
```

```python
y_test
```

# **linear regression**

```python
modellinear=Pipeline([("tfidf",TfidfVectorizer()),('clf',LinearSVC(dual=True))]) 
modellinear.fit(x_train,y_train)
```

```python
modellinear.score(x_test,y_test)
```

# **import model as piclke**

```python
modellinear.predict(x_test.head(1))
```

```python
pickle.dump(modellinear,open("model.pickle","wb"))
```

# **bagclassi modellinear**

```python
bag_classiline=BaggingClassifier(LinearSVC(dual=True),n_estimators=6)
modellinear=Pipeline([("tfidf",TfidfVectorizer()),('clf',bag_classiline)]) 
modellinear.fit(x_train,y_train)
```

```python
modellinear.score(x_test,y_test)
```

# **KNeighbord**

```python
model=Pipeline([("tfidf",TfidfVectorizer()),('clf',KNeighborsClassifier(n_neighbors=5))]) 
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
bag_classi=BaggingClassifier(KNeighborsClassifier(n_neighbors=5),n_estimators=3)
```

```python
modelbag=Pipeline([("tfidf",TfidfVectorizer()),('clf',bag_classi)])
modelbag.fit(x_train,y_train)
```

```python
modelbag.score(x_test,y_test)
```

# **test with decission tree**

```python
decision=DecisionTreeClassifier(max_depth=10)
modelbagdeciss=Pipeline([("tfidf",TfidfVectorizer()),('clf',decision)])
modelbagdeciss.fit(x_train,y_train)
```

```python
modelbagdeciss.score(x_test,y_test)
```
