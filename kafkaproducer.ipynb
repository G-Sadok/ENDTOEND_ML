```python
from kafka import KafkaProducer
import json,requests,time
```

```python
producer = KafkaProducer(
    bootstrap_servers=['localhost:9092'],
    value_serializer=lambda v: json.dumps(v).encode('utf-8')  # Serialize data to JSON
)

def send_tweet(film):
    try:
        while True:
            requet=requests.get('https://www.reddit.com/r/netflix.json',headers={'User-Agent':"Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:145.0) Gecko/20100101 Firefox/145.0"}).json()
            for data in requet['data']['children']:
                producer.send('predict-model',data)
                time.sleep(10)
            time.sleep(60)
    except Exception as e:
        print(f'error key {e}')
```

```python
send_tweet('netflix')
```
